```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

#read dataset
train_data = pd.read_csv("train_dataset_decision_tree.csv")
test_data = pd.read_csv("test_dataset_decision_tree.csv")

#shape of dataset
print("Shape of training dataset: ", train_data.shape)
print("Shape of testing dataset: ", test_data.shape)

#we need to predict the value of missing target variable in test data
#target variable - Survived

#segregating dataset into input(x) and output(y) datasets

#training dataset
x_train = train_data.drop(columns=['Survived'], axis = 1)
y_train = train_data['Survived']

#testing dataset
x_test = test_data.drop(columns=['Survived'], axis = 1)
y_test = test_data['Survived']

#creating decision tree model
model = DecisionTreeClassifier()

#training it
model.fit(x_train,y_train)

#depth of tree
print("\nDepth of decision tree: ", model.get_depth())

#predict target on training dataset
train_predictions = model.predict(x_train)
print("\n'Survived' prediction on training dataset: ", train_predictions)

#accuracy score train dataset
accuracy_train = accuracy_score(y_train, train_predictions)
print("\nAccuracy score on train dataset: ", accuracy_train)

#predict target on training dataset
test_predictions = model.predict(x_test)
print("\n'Survived' prediction on testing dataset: ", test_predictions)

#accuracy score testing dataset
accuracy_test = accuracy_score(y_test, test_predictions)
print("\nAccuracy score on test dataset: ", accuracy_test)
```
